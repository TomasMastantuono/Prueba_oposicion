```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ParÃ¡metros del sistema
L, m, g, k = 10, 1, 9.8, 1  # --> alpha = mg/k
w = np.sqrt(k / (2*m))
x_eq = m*g / k

# Defino las soluciones
x = lambda t: (L - (m * g)/k) * np.cos(w * t) + (m*g / k)
t = np.linspace(0, 30, 10_000)
```

```python
plt.figure()

#------------------
# Figura principal
#------------------
plt.plot(t, x(t), '-', color = 'orange')

plt.xticks([]), plt.yticks([])
plt.axhline(x_eq, linestyle = '--', color = 'blue')

plt.ylabel(r'$x(t)$', fontsize = 15), plt.xlabel(r'$t$', fontsize = 15)

#----------
# Texto
#----------

texto = r'$x_{eq}$'
plt.text(0, x_eq + 1e-2, texto,
         fontsize = 20, color = 'blue')

plt.savefig('src/solucion_xt.pdf', bbox_inches = 'tight')
plt.show()
```
